### 정규식 - re() 모듈

### re.compile(pattern)
- 특정 기호를 정규 표현식 객체로 만들어 준다. 
- re 라이브러리를 사용하려면 정규 표현식 패턴을 매번 작성해야 하는데, 이 함수를 사용해 패턴을 컴파일하면 필요할 때마다 사용할 수 있다.

```python
pattern = ' \W+'
re_pattern = re.compile(pattern)
```

### re.search(pattern, string)
- 해당 문자열에서 정규표현식에 해당하는 첫 부분을 찾는다.
- "(\w+)" 패턴 : 탭, 줄바꿈, 공백이 아닌 문자를 모두 찾는다.

```python
re.search( "(\w+)", "wow, it is awesome" )
```

- a가 들어간 문자를 찾는다.

```python
re.search( "(a)", "wow, it is awesome" )
```

### re.split(pattern, string)
- 해당 문자열에서 특정 패턴으로 문자열을 나눠서 리스트로 만든다.
- (\w+) 패턴 : 문자와 숫자가 아닌 것, 공백을 기준으로 문장을 나눔

```python
re.split("(\w+)", "wow, it is world of word")
```

### re.sub(pattern, repl, string)
- 문자열(string)에서 특정 패턴(pattern)을 만족시키는 문자를 사용자가 정의한 문자(repl)로 치환
- "\d" 패턴 : 숫자를 지정하는 패턴, 숫자를 찾아 "number"로 치환

```python
re.sub("\d", "number", "7 candy" )
```

```python
re.sub("[0-9]", "number", "7 candy" )
```

```python
re.sub("[A-Za-z]", " ", "7 candy" )
```

### Beautiful soup 사용

```python
from bs4 import BeautifulSoup
```

```python
string = '<body> 이 글은 Beautiful soup 라이브러리를 사용하는 방법에 대한 글입니다. <br> </br> 라이브러리를 사용하면 쉽게 HTML 태그를 제거할 수 있습니다.</body>'

string = BeautifulSoup(string,"html5lib").get_text() # HTML 태그를 제외한 텍스트만 가져온다

print(string) # 텍스트 확인
```
