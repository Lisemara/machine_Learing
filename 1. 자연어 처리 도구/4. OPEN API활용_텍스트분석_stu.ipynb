### 제목, 링크만 뽑아서 엑셀에 넣기

```python
#!pip install openpyxl
```

```python
import requests
from urllib.parse import urlparse
import pandas as pd

# 검색결과 저장
lc_title = []
lc_link = []

# 함수 선언
def get_api_result(keyword, display, start) :
    url = "https://openapi.naver.com/v1/search/blog?query=" + keyword + "&display=" + str(display) + "&start=" + str(start)
    result = requests.get(urlparse(url).geturl(),
                        headers = {"X-Naver-Client-Id":"frBMA0JHYp1zATqntdUJ",
                                  "X-Naver-Client-Secret":"CzMu_Uce6r"})
    return result.json()

def call_and_print(keyword, page):
    json_obj = get_api_result(keyword, 100, page)
    for item in json_obj['items'] :
        title = item['title'].replace("<b>", "").replace("</b>", "")
        link = item['link']
        
        # 결과를 리스트에 넣기
        lc_title.append(title)
        lc_link.append(link)

# 메인 코드    
keyword="신종코로나"
call_and_print(keyword, 1) # 1-100


# pandas를 이용하여 csv, excel로 저장하기
df = pd.DataFrame()

df['title']=lc_title
df['link']=lc_link

df.to_csv("naver_blog_output.csv",encoding="utf-8-sig", index=False)
df.to_excel("naver_blog_output.xlsx", index=False)
```

### 단어 빈도 그래프 및 워드 클라우드 그리기 
#### title만 텍스트 파일로 저장하기 

```python
import requests
from urllib.parse import urlparse
import pandas as pd


def get_api_result(keyword, display, start) :
    url = "https://openapi.naver.com/v1/search/blog?query=" + keyword + "&display=" + str(display) + "&start=" + str(start)
    result = requests.get(urlparse(url).geturl(),
                        headers = {"X-Naver-Client-Id":"frBMA0JHYp1zATqntdUJ",
                                  "X-Naver-Client-Secret":"CzMu_Uce6r"})
    return result.json()

def call_and_print(keyword, page):
    tot_title = ""     #검색결과 저장
    json_obj = get_api_result(keyword, 100, page)
    for item in json_obj['items'] :
        title = item['title'].replace("<b>", "").replace("</b>", "")  

        # 결과를 텍스트로 만들기
        tot_title += title
        
    return tot_title
    
keyword="신종코로나"
title_100 = call_and_print(keyword, 1) # 1-100
print(title_100)
```

### 오픈 API로 네이버 뉴스 데이타 수집하기

```python
# 오픈 API로 데이타 수집하기
import requests
from urllib.parse import urlparse
import pandas as pd


def get_api_result(keyword, display, start) :
   #url = "https://openapi.naver.com/v1/search/blog?query=" + keyword + "&display=" + str(display) + "&start=" + str(start)
    url = "https://openapi.naver.com/v1/search/news?query=" + keyword + "&display=" + str(display) + "&start=" + str(start)    
    result = requests.get(urlparse(url).geturl(),
                        headers = {"X-Naver-Client-Id":"frBMA0JHYp1zATqntdUJ",
                                  "X-Naver-Client-Secret":"CzMu_Uce6r"})
    return result.json()

def call_and_print(keyword, page):
    tot_title = ""     #검색결과 저장
    json_obj = get_api_result(keyword, 100, page)
    for item in json_obj['items'] :
        title = item['title'].replace("<b>", "").replace("</b>", "")  

        # 결과를 텍스트로 만들기
        tot_title += title
        
    return tot_title
    
keyword="신종코로나"
title_100 = call_and_print(keyword, 1) # 1-100
print(title_100)
```

### 형태소 분석

```python
# 형태소 분석기
from konlpy.tag import Okt
okt = Okt()
```

```python
# 명사만 추출 - 글자 길이가 2 이상인 단어만 추출
tokens_ko = okt.nouns(title_100) 
tokens_ko = [ ko for ko in tokens_ko if len(ko) > 1 ]
tokens_ko[:10]
```

```python
# nltk
ko = nltk.Text(tokens_ko, name='신종코로나')
ko.vocab().most_common(20)
```

```python
ko.similar('진자')
```

### 워드 클라우드

```python
data = ko.vocab().most_common(50)

wordcloud = WordCloud(font_path= "c:/Windows/Fonts/malgun.ttf",
                      relative_scaling = 0.2,
                      #stopwords=STOPWORDS,
                      background_color='white',
                      ).generate_from_frequencies(dict(data))
plt.figure(figsize=(12,7))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

### 원하는 이미지를 이용한 워드 클라우드 그리기

```python
from wordcloud import ImageColorGenerator
import numpy as np

mask = np.array(Image.open('heart.jpg'))
image_colors = ImageColorGenerator(mask)
```

```python
data = ko.vocab().most_common(200)

wordcloud = WordCloud(font_path= "c:/Windows/Fonts/malgun.ttf",
               relative_scaling = 0.1, mask=mask,
               background_color = 'white',
               min_font_size=1,
               max_font_size=100).generate_from_frequencies(dict(data))

default_colors = wordcloud.to_array()

plt.figure(figsize=(10,7))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
